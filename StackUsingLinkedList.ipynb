```c++
#include <iostream>
#include <assert.h> 
using namespace std;
```

```c++
struct node
{
    int data;
    struct node *next;
};
```

```c++
void Push(struct node ** TOP, int data)
{
    struct node * temp = new node;
    temp->data = data;
    temp->next = *TOP;
    *TOP = temp;
    cout<<"pushed :"<<(*TOP)->data<<endl;
}
```

```c++
int pop (struct node ** TOP)
{
    int ret = -1;
    assert(*TOP != NULL);

    struct node * temp = *TOP;
    ret = temp->data;
    *TOP=(*TOP)->next;
    delete temp;        

    cout<<"popped :"<<ret<<endl;
    return ret;
}
```

```c++
int isEmpty(struct node * TOP)
{
 return TOP == NULL;
}
```

```c++
void show(struct node * TOP)
{ 
    struct node * temp = TOP;
    while(!isEmpty(temp))
    {
        cout<<temp->data<<" ";
        temp = temp->next;
    }
}
```

```c++
struct node * stack = NULL;
```

```c++
isEmpty(stack)
```

```c++
Push(&stack,1);
Push(&stack,2);
Push(&stack,3);
Push(&stack,4);
Push(&stack,5);
```

```c++
isEmpty(stack)
```

```c++
show(stack)
```

```c++
pop(&stack);
pop(&stack);
pop(&stack);   
```

```c++
show(stack)
```
