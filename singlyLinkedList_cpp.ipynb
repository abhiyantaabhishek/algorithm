```c++
#include <iostream>
#include <assert.h> 
using namespace std;
```

```c++
template <class T>
struct node
{
    T data;
    struct node *next;
};
template <class T>
class linkedList
{
public:
    struct node<T> * head = NULL, *tail = NULL;

};
```

```c++
template<class T>
void insert_start(linkedList<T> & List,int val)
{
    struct node<T> * curr = new node<T>;
    curr->data = val;
    if (List.head == NULL)
    {
        curr->next = NULL;
        List.head = curr;
        List.tail = curr;
    }
    else
    {
        curr->next = List.head;
        List.head = curr;
    }
}
```

```c++
template<class T>
void insert_last(linkedList<T> & List , int val)
{
    struct node<T> * newNode = new node<T>;
    struct node<T> * curr = List.head;
    struct node<T> * prev = NULL;
    newNode->data = val;
    while (curr->next != NULL)
    {
        prev = curr;
        curr = curr->next;
    }

    curr->next = newNode;
    newNode->next = NULL;
    List.tail = newNode;

}
```

```c++
template<class T>
void insert_pos(linkedList<T> & List, int val, int pos) // function to insert any position, pos follows 0 indexing
{
    int invalidpos = 0;
    if (pos >= 0)
    {


        int iter = 0;
        struct node<T> * newNode = new node<T>;
        struct node<T> * curr = List.head;
        struct node<T> * prev = NULL;
        newNode->data = val;
        for (iter = 0; iter < pos; iter++)
        {
            if (curr == NULL)
            {
                invalidpos = 1;
                break;

            }
            prev = curr;
            curr = curr->next;
            //assert(curr == NULL);
        }
        if (invalidpos == 0)
        {
            if (iter == 0)
            {
                if (curr == NULL)
                {
                    List.head = newNode;
                    newNode->next = NULL;

                }
                else
                {
                    newNode->next = curr;
                    List.head = newNode;
                }

            }
            else if (curr == NULL)
            {
                prev->next = newNode;
                newNode->next = NULL;
                List.tail = newNode;

            }
            else
            {
                newNode->next = prev->next;
                prev->next = newNode;
            }
        }
        else
        {
            cout << "\ninsert_pos: crossed Max size";
        }
    }
    else
    {
        cout << "\ninsert_pos: crossed Min size";
        return;
    }

}
```

```c++
template <class T>
void delete_pos(linkedList<T> & List,  int pos)// function to delete from linkedList, any position, pos follows 0 indexing
{
    int overFlow = 0;

    struct node<T> * prev = NULL;
    struct node<T> *curr = List.head;
    if (pos >= 0)
    {
        int iter = 0;
        for (iter = 0; iter < pos; iter++)
        {

            prev = curr;
            curr = curr->next;
            if (curr == NULL)
            {
                overFlow = 1;
                break;  // break if loop crossed size 
            }
        }
        if (overFlow == 0)
        {
            if (iter == 0) // delete 0th index and chnage head pointer 
            {
                List.head = curr->next;
                delete curr;

            }
            else if (curr->next == NULL) // delete last element and change the tail pointer
            {
                prev->next = NULL;
                List.tail = prev;
                delete curr;

            }
            else  // delte the element between first and last element 
            {
                prev->next = curr->next;
                delete curr;

            }
        }
        else
        {
            cout << "\ndelete_pos: Index higher than Max size can not be deleted"<<endl;
        }
    }
    else
    {
        cout << "\ndelete_pos: Index can not be negative" << endl;
    }


}
```

```c++
template<class T>
void insert_between_first_and_last(linkedList<T> & List, int val,int pos)// can insert only between first and last element, pos follows 0 indexing
{
    struct node<T> * newNode = new node<T>;
    struct node<T> * curr = List.head;
    struct node<T> * prev = NULL;
    newNode->data = val;

    for (int iter = 1; iter <= pos; iter++)
    {
        prev = curr;
        curr = curr->next;
    }
    newNode->next = prev->next;
    prev->next = newNode;
}
```

```c++
template<class T>
void display(linkedList<T> & List) // fucntion to display the linkedList 
{
    struct node<T> * curr = List.head;

    while (curr != NULL)
    {
        cout << curr->data << ", ";
        curr = curr->next;
    }
}
```

```c++
int __main()
{
    linkedList<int> List;
    insert_pos(List, 3, 0); // insert 3 at pos 0 ; content of list : 3			;posttion is 0 indexing 
    insert_pos(List, 1, 0);// insert 1 at pos 0 ; content of list : 1 3			;posttion is 0 indexing 
    insert_pos(List, 2, 1);// insert 2 at pos 1 ; content of list : 1 2 3		;posttion is 0 indexing 
    insert_pos(List, 4, 3);// insert 4 at pos 4 ; content of list : 1 2 3 4		;posttion is 0 indexing 
    display(List);  // displays 1 2 3 4
    delete_pos(List, 3);// deltes pos 3 element ; content of list : 1 2 3 		;posttion is 0 indexing  
    cout << endl;
    display(List);		// diplays 1 2 3
    insert_pos(List, 8, 5);// can not insert as pos 5 as pos 3 and 4 is not filled 
    insert_pos(List, 8, -1); // can not insert at negaive index
    return 0;
}
```
